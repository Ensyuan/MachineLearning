```python
#----------------------------------
# Toy Datasets (Iris 鳶尾花)
#----------------------------------


from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
print(type(iris.data))
print(iris.feature_names)

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_df['species'] = iris.target 

# 將品種加入 data frame
```

```python
#----------------------------------
# 建立線性迴歸分析模型
#----------------------------------
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

temp = np.array([29, 28, 34, 31, 25, 29, 32, 31, 24, 33, 25, 31, 26, 30])
iced_tea_sales = np.array([77, 62, 93, 84, 59, 64, 80, 75, 58, 91, 51, 73, 65, 84])

lm = LinearRegression()
lm.fit(np.reshape(temp, (len(temp), 1)), np.reshape(iced_tea_sales, (len(iced_tea_sales),1)))

print('coef: ',lm.coef_)

print('intercept: ',lm.intercept_)
```

```python
#----------------------------------
# 利用線性迴歸分析模型預測
#----------------------------------
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

temp = np.array([29, 28, 34, 31, 25, 29, 32, 31, 24, 33, 25, 31, 26, 30])
iced_tea_sales = np.array([77, 62, 93, 84, 59, 64, 80, 75, 58, 91, 51, 73, 65, 84])

lm = LinearRegression()
lm.fit(np.reshape(temp, (len(temp), 1)), np.reshape(iced_tea_sales, (len(iced_tea_sales),1)))

# new Tepm
to_be_predicted = np.array([30])
predicted_sales = lm.predict(np.reshape(to_be_predicted, (len(to_be_predicted), 1)))

# 預測冰紅茶銷量
print('Predict Sale: ',predicted_sales)
```

```python
#----------------------------------
# 線性迴歸視覺化
#----------------------------------
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

temp = np.array([29, 28, 34, 31, 25, 29, 32, 31, 24, 33, 25, 31, 26, 30])
iced_tea_sales = np.array([77, 62, 93, 84, 59, 64, 80, 75, 58, 91, 51, 73, 65, 84])

lm = LinearRegression()
lm.fit(np.reshape(temp, (len(temp), 1)), np.reshape(iced_tea_sales, (len(iced_tea_sales),1)))

# new Tepm
to_be_predicted = np.array([30])
predicted_sales = lm.predict(np.reshape(to_be_predicted, (len(to_be_predicted), 1)))

# 視覺化
plt.scatter(temp, iced_tea_sales, color='black')
plt.plot(temp, lm.predict(np.reshape(temp, (len(temp), 1))), color='blue', linewidth=3)
plt.plot(to_be_predicted, predicted_sales, color='red', marker='^', markersize=10)
# plt.xticks(())
# plt.yticks(())
plt.show()
```
